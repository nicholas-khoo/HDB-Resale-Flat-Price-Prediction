```python
# import the necessary libraries for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

%matplotlib inline
```

```python
# import cleaned train dataset and assign to a dataframe
df_train = pd.read_csv('../datasets/train_cleaned.csv', index_col=0, low_memory=False)

# set max_rows and max_columns to enable readability
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df_train.head(2)
```

```python
# create a new dataframe with only 'resale_price'
df_cal_mean = df_train[['resale_price']]
```

```python
# compute mean and save to a variable
mean_value = df_cal_mean['resale_price'].mean()
mean_value
```

```python
# import the test dataset to obtain 'id' values
df_test = pd.read_csv('../datasets/test.csv')
```

```python
df_test.head(2)
```

```python
# extract 'id' values and assign to a new dataframe
df_kaggle_submit = df_test[['id']]
```

```python
# check the number of rows within dataframe
df_kaggle_submit.shape
```

```python
# add mean values to the dataframe with column name 'Predicted'
df_kaggle_submit['Predicted'] = mean_value
```

```python
df_kaggle_submit.shape
```

```python
# export for submission to kaggle
df_kaggle_submit.to_csv('../datasets/kaggle_baseline_submission.csv', index=False)
```
